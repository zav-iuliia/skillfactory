```python
from duckduckgo_images_api import search
import requests as r
import shutil
from tqdm.auto import tqdm
import warnings

warnings.filterwarnings("ignore")
import logging

logger = logging.getLogger()
logger.setLevel(logging.CRITICAL)
```

```python
def download_file(url, filename):
    res = r.get(url, stream=True)
    with open(filename, "wb") as out_file:
        shutil.copyfileobj(res.raw, out_file)
    del res
```

```python
results = search("Upupa Epops")#bird name
```

```python
BASE_FOLDER = "/Users/iuliia/Desktop/Birds/"
folder = BASE_FOLDER + "Upupa_Epops/"
for res in tqdm(results["results"]):
    url = res["image"]
    filename = folder + url.split("/")[-1].split("?")[0]
    try:
        download_file(url, filename)
    except Exception as e:
        print(f"Problem with {url} - {filename}: {e}")
```
